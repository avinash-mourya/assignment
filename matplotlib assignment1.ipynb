```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
iris
```

```python
pca = PCA(n_components=3)
X = pca.fit_transform(iris["data"])
Y = iris["target"]
```

```python
X
```

```python
fig=plt.figure(figsize=(8,10))
ax3d = Axes3D(fig=fig,elev=-150, azim=110)
ax3d .scatter(X[:, 0], X[:, 1], X[:, 2], c=Y,
           cmap=plt.cm.Set1, edgecolor='k', s=40)
ax3d.set_title("iris 3D visulization")
ax3d.set_xlabel("pca1")
ax3d.w_xaxis.set_ticklabels([])
ax3d.set_ylabel("pca2")
ax3d.w_yaxis.set_ticklabels([])
ax3d.set_zlabel("pca3")
ax3d.w_zaxis.set_ticklabels([])
```
